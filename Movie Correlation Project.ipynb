```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots


# Read in the data
df = pd.read_csv(r'C:\Users\nahur\Documents\Python\movies.csv')
```

```python
df.head()
```

```python
# Checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Column data types

df.dtypes
```

```python
# Changing data type of columsn 
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
df
```

```python
# Correct the year column 
df['yearcorrect'] = df['released'].astype(str).str[:4]
df
```

```python
df.sort_values(by = ['gross'], inplace=False, ascending=False)
```

```python
# Check for duplicates
df.drop_duplicates()
```

```python
df.head()
```

```python
# Is "budget" and "company" correlated with "gross"
```

```python
# Scatter plot with Budget vs Gross Revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Revenue')
plt.xlabel('Gross')
plt.ylabel('Budget')

plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Find the correlation (for numeric fields)

df.corr(numeric_only = True) # Peason
```

```python
correlation_matrix = df.corr(numeric_only = True)

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix')
plt.xlabel('Categories')
plt.ylabel('Categories')

plt.show()
```

```python
# Look at company 
# Convert string to numeric 

df_num = df

for col_name in df_num.columns:
    if(df_num[col_name].dtype == 'object'):
        df_num[col_name] = df_num[col_name].astype('category')
        df_num[col_name] = df_num[col_name].cat.codes
        
df_num
```

```python
correlation_matrix = df_num.corr(numeric_only = True)

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix')
plt.xlabel('Categories')
plt.ylabel('Categories')

plt.show()
```

```python
df_num.corr()
```

```python
# Unstacking

correlation_matrix = df_num.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
